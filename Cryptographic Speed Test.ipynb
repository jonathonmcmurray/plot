### Comparing the speed of kdb+ hashing functions

Taking `.Q.sha1` (available in v3.6+) and comparing speedwise to `md5`, for different lengths of input strings

Using `plot` to create simple visualisations

```q
.utl.require"plot"
```

```q
//demonstrate hash functions
md5"hello world"
.Q.sha1"hello world"
```

```q
//define a simple timing function
s:{[f;x]st:.z.p;do[10000;f@x];.z.p-st}
//define a test corpus
c:t!(t:1+til 1000)#\:.Q.an
```

```q
//run timing for each hash function
sh:s[.Q.sha1]each c
md:s[md5]    each c
```

## Plotting results

```q
plot`plot`x`y`xlabel`ylabel`title`grid!(`line;key sh;value sh;"String Length";"Time Taken";".Q.sha1 timing";1b)
```

```q
plot`plot`x`y`xlabel`ylabel`title`grid!(`line;key md;value md;"String Length";"Time Taken";"md5 timing";1b)
```
