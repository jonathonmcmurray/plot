## `plot` - plotting module for qutil

Simple wrapper around matplotlib. Includes [`importmatplotlib.q`](https://github.com/KxSystems/embedPy/blob/master/examples/importmatplotlib.q) from kx. Requires [embedPy](https://github.com/KxSystems/embedPy), works nicely with [jupyterq](https://github.com/KxSystems/jupyterq)

Built for [qutil](https://github.com/nugend/qutil), hosted on [Anaconda Cloud](https://anaconda.org/jmcmurray/repo)

Accepts dictionaries with some or all args. Little error handling to prevent errors with missing required args. Suitable only for the most basic of plots, for anything with more complex requirements more direct usage of matplotlib will be required

Setup:

* Install kdb, embedPy & jupyterq. Easiest way is via [Anaconda](https://www.anaconda.com/download/):

    `conda install -c kx jupyterq  #other packges installed via dependency`
    
    
* Install necessary q modules. Again, via Anaconda (reQ used for downloading datasets):

    `conda install -c jmcmurray req plot`
    
    
* Run Jupyter in directory containing notebook:

    `jupyter notebook`
    
Without jupyter, also works from console; display locally in Qt window or save to file

```q
.utl.require"plot"   //import the plot module
```

### `line` plot

Replicating the example from kx example notebook at https://github.com/KxSystems/jupyterq/blob/master/kdb%2BNotebooks.ipynb

```q
pi:2*asin 1
x:til[200]%100
y:1+sin 2*pi*x
plot`plot`x`y`xlabel`ylabel`title`grid!(`line;x;y;"X values";"Y Values";"A pyplot chart created in q";1b)
```

Example with multiple series

```q
x:til[200]%100
y:`liner`quadratic`cubic!(x;x*x;x*x*x)
plot`plot`x`y`xlabel`ylabel`title`grid!(`line;x;y;"X values";"Y Values";"A pyplot chart created in q";1b)
```

### `pie` plot

Partially replicating matplotlib example pie chart from https://matplotlib.org/2.0.2/examples/pie_and_polar_charts/pie_demo_features.html

```q
l:`Frogs`Hogs`Dogs`Logs
s:15 30 45 10
plot`plot`labels`sizes`title!(`pie;l;s;"An example pie chart from q")
```
